## This notebook presents the accuracy of a two layer neural network changes in relationship with the size of tensors in the first and second layer. 

```python
import tensorflow as tf
```

```python
import matplotlib.pyplot as plt
```

```python
 (x_train, y_train),(x_test, y_test) = tf.keras.datasets.mnist.load_data()
    
x_train = x_train[0:1000]
y_train = y_train[0:1000]
```

```python

plt.imshow(x_train[1])
plt.show()
```

```python
def train_mnist(x_train, y_train, x,y,epo, switich):
    # YOUR CODE SHOULD START HERE
    x_train = x_train/255;
    # YOUR CODE SHOULD END HERE
    model = tf.keras.models.Sequential([
        
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(x,activation=tf.nn.relu),
        tf.keras.layers.Dense(y,activation=tf.nn.relu),
        tf.keras.layers.Dense(10,activation = tf.nn.softmax)
    ]);

    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy']);
    
    # model fitting
    history = model.fit(  # YOUR CODE SHOULD START HERE
    x_train,y_train,epochs= epo
              # YOUR CODE SHOULD END HERE
    );
    # model fitting
    
    if switich == True:
        model.evaluate(x_test,y_test)
    return history.history['accuracy'][-1]
```

```python
num = train_mnist (x_train, y_train,360,64,1,False)
```

```python
x_a = x_b = list(range(10,1000,25))
len(x_a)
```

```python
counter_1 = 0
counter_2 = 0
for i in x_a:
    for j in x_b:
        num = train_mnist(x_train, y_train,i,j,1,False) ;
        k[counter_1,counter_2] = num ;
        counter_1 = counter_1 + 1 ;
    counter_1 = 0
    counter_2 = counter_2 + 1; 
        
```
